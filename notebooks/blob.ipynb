```python
# Download the PDF file from Azure Blob Storage (streaming)
pdf_stream = blob_client.download_blob().content_as_bytes()
pdf_stream
```

```python
# Read the PDF content using PyMuPDF directly from the stream
def read_pdf_content_from_stream(pdf_stream):
    try:
        # Create a PyMuPDF Document object from the PDF stream
        pdf_document = fitz.open(stream=pdf_stream, filetype="txt")

        # Initialize an empty string to store the text content
        text_content = ""

        # Iterate through pages and extract text
        for page_number in range(pdf_document.page_count):
            # Get the page
            page = pdf_document[page_number]

            # Extract text from the page
            text_content += page.get_text()

        return text_content

    except Exception as e:
        # Handle exceptions, such as invalid PDF format or corrupted file
        print(f"Error: {e}")
        return None

# Read the PDF content directly from the stream
pdf_text_content = read_pdf_content_from_stream(pdf_stream)

# Print or use the extracted text content as needed
print(pdf_text_content)
```
